```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import os

# Typography / plot style (legends are mechanism names only)
mpl.rc('font', family='serif', serif=['Times New Roman'])
mpl.rcParams.update({
    'text.usetex': False,
    'font.size': 16,
    'axes.titlesize': 20,
    'axes.labelsize': 20,
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 20,
    'figure.titlesize': 20,
})
mpl.rcParams['text.usetex'] = False
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 30
def plot_phase1_from_csv(
    csv_path,
    plot_dir=None,           # e.g., "plots"
    pdf_name=None            # e.g., "mseVSphase1_from_csv.pdf" (optional)
):
    """
    Load a CSV produced by sweep_vs_phase1(...) and plot with the exact same style.
    Expected columns: 'phase1_pct' and any subset of ['SPL','RS+FD','RS+RFD','Corr-RR'].
    """

    # ---- load ----
    df = pd.read_csv(csv_path)
    if "phase1_pct" not in df.columns:
        raise ValueError("CSV must contain a 'phase1_pct' column.")
    x_vals = df["phase1_pct"].astype(int).tolist()

    # Which series exist in the CSV
    keys = [k for k in ["SPL", "RS+FD", "RS+RFD", "Corr-RR"] if k in df.columns]
    if not keys:
        raise ValueError("CSV has no mechanism columns. Expected one of: SPL, RS+FD, RS+RFD, Corr-RR.")

    # ---- plot (exact same style) ----
    plt.figure(figsize=(10, 8))

    if "SPL" in keys:
        plt.plot(x_vals, df["SPL"].values,    '-o', linewidth=3, markersize=16, label='SPL')
    if "RS+FD" in keys:
        plt.plot(x_vals, df["RS+FD"].values,  '-s', linewidth=3, markersize=16, label='RS+FD')
    if "RS+RFD" in keys:
        plt.plot(x_vals, df["RS+RFD"].values, '-^', linewidth=3, markersize=16, label='RS+RFD')
    if "Corr-RR" in keys:
        plt.plot(x_vals, df["Corr-RR"].values,'-D', linewidth=3, markersize=16, label='Corr-RR')

    plt.xlabel('Users in Phase I (%)', fontsize=40)
    plt.ylabel('MSE', fontsize=40)
    plt.xticks(x_vals, labels=[f"{x}" for x in x_vals])
    plt.tick_params(axis='both', which='major', labelsize=30)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))

    # Legend at top center, 2 columns, same spacing/layout
    plt.legend(
        fontsize=35,
        loc='upper center',
        ncol=2,
        frameon=True,
        edgecolor='black',
        bbox_to_anchor=(0.5, 1.33),
        labelspacing=0.01,
        columnspacing=0.8,
        handletextpad=0.8,
        handlelength=2
    )

    plt.tight_layout()
    plt.subplots_adjust(top=0.82)

    # ---- save (optional) ----
    if plot_dir is not None:
        os.makedirs(plot_dir, exist_ok=True)
        if pdf_name is None:
            base = os.path.splitext(os.path.basename(csv_path))[0]
            pdf_name = f"{base}.pdf"
        out_path = os.path.join(plot_dir, pdf_name)
        plt.savefig(out_path, format="pdf")

    plt.show()
```

```python
def plot_rho_from_csv(
    csv_path,
    plot_dir=None,           # e.g., "plots"
    pdf_name=None            # e.g., "mseVSrho_from_csv.pdf" (optional)
):
    """
    Load a CSV produced by a rho-sweep (fixed epsilon, varying correlation) and
    plot with the exact same style you used before.
    Expected columns:
      - 'rho'  (x-axis)
      - any subset of ['SPL','RS+FD','RS+RFD','Corr-RR'] (y-series)
    """

    # ---- load ----
    df = pd.read_csv(csv_path)

    # pick x column (rho)
    if "rho" not in df.columns:
        # fallback names in case your CSV used another header
        for alt in ("correlation", "corr"):
            if alt in df.columns:
                df = df.rename(columns={alt: "rho"})
                break
    if "rho" not in df.columns:
        raise ValueError("CSV must contain a 'rho' column (or 'correlation'/'corr').")

    # sort by rho (safety)
    df = df.sort_values("rho").reset_index(drop=True)
    x_vals = df["rho"].astype(float).tolist()

    # Which series exist in the CSV
    keys = [k for k in ["SPL", "RS+FD", "RS+RFD", "Corr-RR"] if k in df.columns]
    if not keys:
        raise ValueError("CSV has no mechanism columns. Expected one of: SPL, RS+FD, RS+RFD, Corr-RR.")

    # ---- plot (exact same style) ----
    plt.figure(figsize=(10, 8))

    if "SPL" in keys:
        plt.plot(x_vals, df["SPL"].values,    '-o', linewidth=3, markersize=16, label='SPL')
    if "RS+FD" in keys:
        plt.plot(x_vals, df["RS+FD"].values,  '-s', linewidth=3, markersize=16, label='RS+FD')
    if "RS+RFD" in keys:
        plt.plot(x_vals, df["RS+RFD"].values, '-^', linewidth=3, markersize=16, label='RS+RFD')
    if "Corr-RR" in keys:
        plt.plot(x_vals, df["Corr-RR"].values,'-D', linewidth=3, markersize=16, label='Corr-RR')

    plt.xlabel(r'$\rho$', fontsize=50)
    plt.ylabel('MSE',     fontsize=40)
    plt.xticks(x_vals, labels=[str(x) for x in x_vals])
    plt.tick_params(axis='both', which='major', labelsize=30)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))

    # Legend at top center, 2 columns, same spacing/layout as before
    plt.legend(
        fontsize=35,
        loc='upper center',
        ncol=2,
        frameon=True,
        edgecolor='black',
        bbox_to_anchor=(0.5, 1.33),
        labelspacing=0.01,
        columnspacing=0.8,
        handletextpad=0.8,
        handlelength=1.8
    )

    plt.tight_layout()
    plt.subplots_adjust(top=0.82)

    # ---- save (optional) ----
    if plot_dir is not None:
        os.makedirs(plot_dir, exist_ok=True)
        if pdf_name is None:
            base = os.path.splitext(os.path.basename(csv_path))[0]
            pdf_name = f"{base}.pdf"
        out_path = os.path.join(plot_dir, pdf_name)
        plt.savefig(out_path, format="pdf")

    plt.show()
```
